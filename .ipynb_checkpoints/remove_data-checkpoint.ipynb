```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import signal
from datetime import datetime
%matplotlib inline
```

#### Import the open/close time data.

```python
df = pd.read_csv(filepath_or_buffer = 'data/chamber_1/all_cham1_close_times.csv', index_col = None)
df.head()
```

```python
df['end_of_run'] = pd.to_datetime(df.year.astype(str) + '-' + df.month.astype(str) + '-' + df.day.astype(str) + ' ' + df.hour.astype(str) + ':' + df['min'].astype(str) + ':' + df.sec.astype(str))
df = df.set_index(pd.DatetimeIndex(df['end_of_run']))
df.drop('end_of_run', axis=1, inplace=True)
```

```python
df = df.resample('1Min')
df = df.dropna()
```

```python
df.head()
```

```python
df['closing'] = pd.to_datetime(df.year.astype(int).astype(str) + '-' + df.month.astype(int).astype(str) + '-' + df.day.astype(int).astype(str) + ' ' + df.open_hour.astype(int).astype(str) + ':' + df['omin'].astype(int).astype(str))
df['closed'] = pd.to_datetime(df.year.astype(int).astype(str) + '-' + df.month.astype(int).astype(str) + '-' + df.day.astype(int).astype(str) + ' ' + df.close_hour.astype(int).astype(str) + ':' + df['cmin'].astype(int).astype(str))
```

```python
df.head()
```

```python
df = df.drop(['year','month','day','hour','min','sec','close_hour','cmin','csec','open_hour','omin','osec','camb', 'close_month', 'close_day'], axis=1)
```

```python
df['open_start'] = df.closing - pd.Timedelta(minutes=10)
df['open_end'] = df.closing - pd.Timedelta(minutes=3)
df['closed_start'] = df.closed + pd.Timedelta(minutes=4)
df['closed_end'] = df.closed + pd.Timedelta(minutes=7)
```

```python
df.head()
```

```python
date1 = []
for idx,row in df.iterrows():
    i = 0
    this_date = row.open_start
    while this_date < row.open_end:
        this_date = row.open_start + pd.Timedelta(minutes=i)
        date1.append(this_date)
        i += 1
```

```python
df_open = pd.DataFrame(index=date1)
```

```python
df_open.head()
```

```python
date1 = []
for idx,row in df.iterrows():
    i = 0
    this_date = row.closed_start
    while this_date < row.closed_end:
        this_date = row.closed_start + pd.Timedelta(minutes=i)
        date1.append(this_date)
        i += 1
```

```python
df_closed = pd.DataFrame(index=date1)
```

#### Import the pod data.

```python
df_R = pd.read_csv(filepath_or_buffer = 'data/chamber_1/all_cham1.csv', parse_dates = [[1,2]], index_col = 0)
df_R.head()
```

```python
df_R = df_R.resample('1Min', how='median')
```

```python
df_R.head()
```

```python
df_R.drop(['UnixTime','Op_voltage', 'Flow_Rate', 'Soil', 'Rain', 'TempOUT', 'RhOUT', 'run#', 'CO2'], axis=1, inplace=True)
```

```python
df_R.head()
```

```python
df_open = pd.concat([df_open, df_R], join='inner', axis=1)
df_closed = pd.concat([df_closed, df_R], join='inner', axis=1)
```

```python
df_closed.head()
```

```python
df_open.to_csv(path_or_buf = 'data/chamber_1/all_cham1_open.csv')
df_closed.to_csv(path_or_buf = 'data/chamber_1/all_cham1_closed.csv')
```
